novamento realiza sweep agora apos a cross validation. retorna o modelo q sera usado futuramente apos augmentations manuais e tentativa de tratamento das classes minoritarias

```python
from ultralytics import YOLO

# Carrega o modelo best do melhor fold, escolhi o 2 
model = YOLO("runs/detect/fold_2_run/weights/best.pt") 
model.to('cuda')  # GPU

print(f"Modelo carregado em: {next(model.model.parameters()).device}")  # Confirma se foi para a GPU
# Realiza o sweep
space = {
    'lr0': (1e-4, 1e-2),
    'momentum': (0.8, 0.95),
    'weight_decay': (1e-5, 1e-2),
    'warmup_epochs': (0.0, 3.0),
    'hsv_h': (0.0, 0.1),
    'hsv_s': (0.0, 0.9),
    'hsv_v': (0.0, 0.9),
    'fliplr': (0.0, 0.5),
    'degrees': (0.0, 20.0),
    'translate': (0.0, 0.2),
    'scale': (0.5, 1.5),
    'shear': (0.0, 2.0)
}

# Executa o tuning com sweep
model.tune(
    data="data_balanced.yaml",
    epochs=8,
    iterations=8,
    imgsz=640,
    batch=8,
    optimizer='SGD',
    name="sweep_final_balanceado",
    space=space  # <-- define aqui o sweep
)
```
